```python
import numpy as np
import pandas as pd
import dabest
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
my_color_palette =  {'Dz 1mg/kg':'#663399','Dz 3mg/kg':'#4AB2D6','Dz 5mg/kg':'#8FB996','Dz 7mg/kg':'#006594','NaCl':'#594157', 'PTZ 20mg/kg':'#c7991a'}
```

```python
# Load Sz data
Sz_df = pd.read_pickle('Mice_data/'+'Sz_df')
cnd_order_bootstrap = ['NaCl', 'Dz 7mg/kg','Dz 5mg/kg','Dz 3mg/kg','Dz 1mg/kg','PTZ 20mg/kg']
```

```python
title= 'Time to Sz across conditions per week bootstrap normalized'
fig1 = plt.figure(title, figsize=(8, 8)).suptitle(title)
gridspec.GridSpec(12, 12)
ax1 = plt.subplot2grid((12, 12), (0, 0), rowspan=12, colspan=12)

mulit_groupe = dabest.load(Sz_df, idx= cnd_order, x="Condition", y="Normalized TTSZ")
mulit_groupe.mean_diff.plot(ax=ax1, swarm_label='Normalized Time to Seizure',custom_palette=my_color_palette)
stats = mulit_groupe.mean_diff.statistical_tests
ax1.axhline(y=1, color='k', linestyle='--', linewidth=0.5)
```

```python
title= 'Time to Sz across conditions barplot normalized'
fig1 = plt.figure(title, figsize=(8, 8)).suptitle(title)
gridspec.GridSpec(12, 12)
cnd_order = ['Dz 7mg/kg','Dz 5mg/kg','Dz 3mg/kg','Dz 1mg/kg','NaCl', 'PTZ 20mg/kg']
ax1 = plt.subplot2grid((12, 12), (0, 0), rowspan=12, colspan=12)
sns.barplot(ax=ax1,data=Sz_df, x="Condition", y="Normalized TTSZ", order=cnd_order,alpha=0.5, palette=my_color_palette)
ax1.set_ylabel("Normalized Time to Seizure")
ax1.axhline(y=1, color='k', linestyle='--', linewidth=0.5)
```

```python
import numpy as np
import pandas as pd
import dabest
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
my_color_palette =  {'NaCl':'#594157','Convulsive PTZ':'#db3021','Control':'#594157','No inj':'#594157','Post-Ictal':'#3287d1'}
```

```python
# Load data 
channel = 'CA1 R'
df_autocorr = pd.read_pickle('Mice_data/'+'Autocorrelation_PTZ_sz_df')
df_var = pd.read_pickle('Mice_data/'+'Variance_PTZ_sz_df')
df_skw = pd.read_pickle('Mice_data/'+'Skewness_PTZ_sz_df')
df_xcorr = pd.read_pickle('Mice_data/'+'Mean Spatial Correlation_PTZ_sz_df')
df_ll = pd.read_pickle('Mice_data/'+'Sum LL_PTZ_sz_df')

# Merge data frames on multiple columns
df_autocorr['Mean Normalized Variance'] = df_var['Mean Normalized Variance']
df_autocorr['Mean Normalized Skewness'] = df_skw['Mean Normalized Skewness']
df_autocorr['Mean Normalized Mean Spatial Correlation'] = df_xcorr['Mean Normalized Mean Spatial Correlation']
df_autocorr['Mean Normalized Sum LL'] = df_ll['Mean Normalized Sum LL']

df_autocorr.to_pickle('Mice_data/'+'Passive_metrics_PTZ_sz_df')
```

```python
df_passive_metrics = pd.read_pickle('Mice_data/'+'Passive_metrics_PTZ_sz_df')
All_sessions = df_passive_metrics['Filename'].unique()
```

```python
metric = 'Autocorrelation'
title = 'Mean ' + metric + ' distance from Sz in'+ channel
fig1 = plt.figure(title, figsize=(16, 8)).suptitle(title)
gridspec.GridSpec(12, 12)
ax1 = plt.subplot2grid((12, 12), (0, 0), rowspan=12, colspan=12)
sns.lineplot(ax=ax1, data=df_passive_metrics, x='Distance from Sz [min]', y='Mean Normalized '+metric,  errorbar=None, zorder=0, hue="Condition", palette=my_color_palette)
for session in All_sessions:
    df = df_passive_metrics[df_passive_metrics['Filename']==session]
    sns.lineplot(ax=ax1, data=df, x='Distance from Sz [min]', y='Mean Normalized '+metric, errorbar=None,zorder=0, lw=0.5, alpha=0.5,hue="Condition", palette=my_color_palette)
ax1.axvline(x=0, color='k', linestyle='--', linewidth=0.5)
ax1.axhline(y=1, color='k', linestyle='--', linewidth=0.5)
ax1.axvspan(0, 1, edgecolor='w', facecolor='w', linestyle="--", alpha=1, lw=0.1, zorder=1)
ax1.set_xlim(left=-32)
ax1.legend([], [], frameon=False)
```

```python
import numpy as np
import pandas as pd
import dabest
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
my_color_palette =  {'Dz 5mg/kg':'#8FB996','NaCl':'#594157', 'PTZ 20mg/kg':'#c7991a'}
```

```python
# Load SP data
SP_df = pd.read_pickle('Mice_data/'+'Mean_SP_CA1 R')
metric = 'Sum LL 250ms'
cnd_order_bootstrap = ['NaCl','Dz 5mg/kg','PTZ 20mg/kg']
```

```python
title = 'Mean Response to first Pulse in CA1 R at 1[v] normalized by block'
fig = plt.figure(title, figsize=(12, 10)).suptitle(title)
ax1 = plt.subplot2grid((12, 12), (0, 0), rowspan=12, colspan=12)
ax1.axhline(y=1, color='k',  linewidth=0.5, alpha=0.8)
multi_groups_unpaired = dabest.load(SP_df,x="Condition", y='Normalized '+metric ,idx=cnd_order_bootstrap)
multi_groups_unpaired.mean_diff.plot(ax=ax1,raw_marker_size=4,swarm_ylim= [0,1.75],contrast_label="Mean difference 95%CI",contrast_ylim= [-0.5,0.5],custom_palette=my_color_palette)
plt.show()
```

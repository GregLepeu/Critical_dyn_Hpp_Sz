```python
import numpy as np
import pandas as pd
import dabest
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
my_color_palette = {'BZD': '#8FB996', 'BL': '#594157'}
```

```python
metrics = ['Autocorrelation','Variance','Skewness','Sum LL']
metric = 'Autocorrelation'

# Load data across animals
Passive_metric_df = pd.read_pickle('Human_data/'+metric+'_human_df')
```

```python
title = metric+' across particpants'
fig2 = plt.figure(title, figsize=(12, 10)).suptitle(title)
ax2 = plt.subplot2grid((12, 12), (0, 0), rowspan=12, colspan=12)
mulit_groupe = dabest.load(Passive_metric_df,idx=['BL','BZD'], x="Condition", y='Normalized '+metric)
mulit_groupe.mean_diff.plot(ax=ax2,swarm_label='Normalized '+metric, contrast_label="Mean difference 95%CI",custom_palette=my_color_palette)

result_Bootstrap = mulit_groupe.mean_diff.statistical_tests
```
